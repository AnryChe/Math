### Task3
Напишите код на Python, реализующий построение графиков:

1.	окружности,

```python
from pylab import rcParams

x = []
y1 = []
y2 = []
r = 500

for i in range(-500, 501):
    x_ = i
    x.append(x_)
    y1.append(math.sqrt(r**2 - (x_)**2))
    y2.append(-math.sqrt(r**2 - (x_)**2))
    
plt.plot(x,y1)
plt.plot(x,y2)
rcParams["figure.figsize"] = 5, 5
plt.xlabel("x")
plt.ylabel("y")
```

2.	эллипса,

```python
x = []
y1 = []
y2 = []
a = 500
b = 300

for i in range(-500, 501):
    x_ = i
    x.append(x_)
    y1.append(math.sqrt((b**2)*(1 - (x_**2)/(a**2))))
    y2.append(-math.sqrt((b**2)*(1 - (x_**2)/(a**2))))
    
plt.plot(x,y1)
plt.plot(x,y2)
plt.axis([-500, 500, -500, 500])
rcParams["figure.figsize"] = 5, 5
plt.xlabel("x")
plt.ylabel("y")
```
